# Wind Stress, Mixing and PAR

```python
def calculate_windstess(windfile):
    wind = pd.read_csv(windfile, sep='\s+', header=None, 
            names=['day', 'month', 'year', 'hour', 'across', 'along'])
    dt = pd.to_datetime(wind.drop(columns=['across', 'along']))
    wind['dt'] = dt
    wind = wind.set_index(dt)

# C_D = 1.0d-3 * (2.70 / UU + 0.142 + 0.0764 * UU)
#stress_u = unow / UU * C_D * rho_atm * UU ** 2
# stress_v = vnow / UU * C_D * rho_atm * UU ** 2
    rho_atm = 1.25
    wind['UU'] = np.sqrt(wind.across**2 + wind.along**2)
    wind['CD'] = 1.0e-3 * (2.70 / wind.UU + 0.142 + 0.0764 * wind.UU)
    wind['stress_u'] = wind.across / wind.UU * wind.CD * rho_atm * wind.UU ** 2
    wind['stress_v'] = wind.along / wind.UU * wind.CD * rho_atm * wind.UU ** 2
    wind['stress'] = wind.CD * rho_atm * wind.UU ** 2

    winddaily = wind.resample('D').mean()
    winddaily = winddaily.drop(columns=['UU', 'CD', 'stress_u', 'stress_v', 'along', 'across',
                                   'year', 'month', 'day', 'hour'])

    winddaily.stress.plot()
    return winddaily
```

```python
! head -13 ../SoG_hindcast/hoff/04/hoff-n04_13Jul2011c.dat
hoffheadings = ['depth', 'temperature', 'salinity', 'sigma-t', 'micro phytoplankton', 
                'nano phytoplankton', 'pico phytoplankton', 'micro zooplankton', 'nitrate', 
                'ammonium', 'silicon', 'DON detritus', 'PON detritus', 'refractory N detritus', 
                'biogenic Si detritus', 'total momentum eddy diffusivity', 
                'total temperature eddy diffusivity', 'total salinity eddy diffusivity', 
                'photosynthetic available radiation', 'u velocity', 'v velocity']
len(hoffheadings)
```

```python
def find_mixing_par(hofffile):
    with open(hofffile) as data:
        mydata = data.read().split()
    syear = int(mydata[89])
    sday = int(mydata[91])
    eyear = int(mydata[95])
    eday = int(mydata[97])
    myarray = np.asarray(mydata[103:], dtype=float)
    numberpoints = myarray.shape[0]/21/82.
    newarray = np.reshape(myarray, (int(numberpoints), 82, 21))

    tstart = datetime.datetime(syear, 1, 1) + datetime.timedelta(sday)
    tend = datetime.datetime(eyear, 1, 1)  + datetime.timedelta(eday - 2)
    dt = [tstart]
    for tt in range(int(numberpoints-1)):
        dt.append(dt[-1] + datetime.timedelta(1))
    
    # Mixing
    fig, axs = plt.subplots(2, 2, figsize=(12, 10))

    colours= axs[0,0].pcolormesh(dt, -newarray[0, :, 0], newarray[:, :, 16].transpose())
    axs[0,0].set_title(hoffheadings[16])
    fig.colorbar(colours, ax=axs[0,0])
    # based on colormesh only use top 15 m of profile, and average
    mixing = newarray[:, :30, 16].mean(axis=1)
    axs[0, 1].plot(dt, mixing)

    # PAR
    colours = axs[1, 0].pcolormesh(dt, -newarray[0, :, 0], newarray[:, :, 18].transpose())
    axs[1, 0].set_title(hoffheadings[18])
    fig.colorbar(colours, ax=axs[1, 0]);
    # surface PAR
    par = newarray[:, 0, 18]
    axs[1, 1].plot(dt, par)

    mixing_par_df = pd.DataFrame({'dt' : dt, 'Av' : mixing, 'PAR': par})

    mixing_par_df = mixing_par_df.set_index('dt')
    return mixing_par_df
```
